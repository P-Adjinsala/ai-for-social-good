## Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.dpi']=110
sns.set_theme(style='whitegrid')
print('Ready.')
```

## 1. Load & Inspect

```python
df = pd.read_csv('../datasets/civic_survey_sample.csv')
print(f'Shape: {df.shape}')
df.info()
```

## 2. Data Cleaning Pipeline

Real survey data always has issues. We handle them systematically.

```python
# Check for missing values
missing = df.isnull().sum()
print('Missing values per column:')
print(missing[missing > 0] if missing.any() else 'None — clean dataset!')

# Check value ranges
print(f'\nAge range: {df.age.min()} – {df.age.max()}')
print(f'Trust score range: {df.trust_in_ai_score.min()} – {df.trust_in_ai_score.max()}')
print(f'Unique regions: {sorted(df.region.unique())}')
```

## 3. Detecting Sampling Bias

A dataset that over-represents certain groups will produce biased models.
Let us check who is in our survey.

```python
fig, axes = plt.subplots(2, 2, figsize=(13, 8))

# Gender
df['gender'].value_counts().plot(kind='bar', ax=axes[0,0], color='steelblue', edgecolor='white')
axes[0,0].set_title('Gender Distribution', fontweight='bold')
axes[0,0].tick_params(axis='x', rotation=0)

# Education
df['education_level'].value_counts().plot(kind='bar', ax=axes[0,1],
    color=sns.color_palette('Greens_d',4), edgecolor='white')
axes[0,1].set_title('Education Level', fontweight='bold')
axes[0,1].tick_params(axis='x', rotation=15)

# Urban vs Rural
df['setting'].value_counts().plot(kind='pie', ax=axes[1,0],
    autopct='%1.1f%%', colors=['#3498db','#e67e22'],
    wedgeprops={'edgecolor':'white'})
axes[1,0].set_title('Urban vs Rural', fontweight='bold')

# Internet access by setting
internet_by_setting = df.groupby('setting')['has_internet_access'].mean() * 100
internet_by_setting.plot(kind='bar', ax=axes[1,1], color=['#e67e22','#3498db'], edgecolor='white')
axes[1,1].set_title('Internet Access Rate by Setting (%)', fontweight='bold')
axes[1,1].tick_params(axis='x', rotation=0)

plt.suptitle('Who Is In Our Dataset?', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 4. Disaggregated Analysis

Always disaggregate data by key dimensions — averages can hide important inequalities.

```python
# AI awareness by setting and gender
awareness = df.groupby(['setting','gender'])['heard_of_ai'].mean().unstack() * 100
awareness.round(1).plot(kind='bar', figsize=(9,5), edgecolor='white',
    color=sns.color_palette('Set2',3))
plt.title('AI Awareness Rate (%) — by Setting & Gender', fontweight='bold')
plt.ylabel('% who have heard of AI')
plt.xlabel('')
plt.xticks(rotation=0)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()

print(awareness.round(1))
```
